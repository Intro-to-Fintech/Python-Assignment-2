```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# List of sector ETF symbols
sector_symbols = ['XLC', 'XLY', 'XLP', 'XLE', 'XLF', 'XLV', 'XLI', 'XLB', 'XLRE', 'XLK', 'XLU']

# Download historical data for each sector
data = yf.download(sector_symbols, start='2023-01-01', end='2024-01-01')

# Calculate daily and weekly returns
daily_returns = data['Adj Close'].pct_change()
weekly_returns = data['Adj Close'].resample('W').ffill().pct_change()

# Plotting daily returns as bar graphs
plt.figure(figsize=(12, 6))
daily_returns.plot(kind='bar', stacked=True, title='Daily Returns of S&P 500 Sectors', xlabel='Date', ylabel='Returns')
plt.legend(sector_symbols)
plt.show()

# Plotting weekly returns as bar graphs
plt.figure(figsize=(12, 6))
weekly_returns.plot(kind='bar', stacked=True, title='Weekly Returns of S&P 500 Sectors', xlabel='Date', ylabel='Returns')
plt.legend(sector_symbols)
plt.show()
```
