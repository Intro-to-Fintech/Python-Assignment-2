```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def get_sector_data():
    # Define S&P 500 sector tickers
    sector_tickers = ["XLC", "XLY", "XLP", "XLE", "XLF", "XLV", "XLI", "XLB", "XLRE", "XLK", "XLU"]

    # Fetch historical data for each sector
    sector_data = yf.download(sector_tickers, start="2023-01-01", end="2024-01-01", interval = "1d")['Adj Close']

    return sector_data

def plot_sector_performance(sector_data):
    # Plot the S&P 500 sector performance
    sector_data.plot(figsize=(12, 8))
    plt.title("S&P 500 Sector Performance")
    plt.xlabel("Date")
    plt.ylabel("Adjusted Close Price")
    plt.legend(sector_data.columns)
    plt.show()

if __name__ == "__main__":
    # Get S&P 500 sector data
    sector_data = get_sector_data()

    # Plot the sector performance
    plot_sector_performance(sector_data)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def get_sector_data():
    # Define S&P 500 sector tickers
    sector_tickers = ["XLC", "XLY", "XLP", "XLE", "XLF", "XLV", "XLI", "XLB", "XLRE", "XLK", "XLU"]

    # Fetch historical data for each sector on a weekly basis
    sector_data = yf.download(sector_tickers, start="2023-01-01", end="2024-01-01", interval="1wk")['Adj Close']

    return sector_data

def plot_sector_performance(sector_data):
    # Plot the S&P 500 sector performance
    sector_data.plot(figsize=(12, 8))
    plt.title("S&P 500 Sector Performance (Weekly)")
    plt.xlabel("Date")
    plt.ylabel("Adjusted Close Price")
    plt.legend(sector_data.columns)
    plt.show()

if __name__ == "__main__":
    # Get S&P 500 sector data on a weekly basis
    sector_data = get_sector_data()

    # Plot the sector performance
    plot_sector_performance(sector_data)
```
